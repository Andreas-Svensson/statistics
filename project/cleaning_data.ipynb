```python
import os
import pandas as pd

# read dataset
file_path = os.path.realpath('../project/assets/iris.csv')
iris = pd.read_csv(file_path)

iris.head()
```

## Renaming Values

The class column includes full genus-species names which is redundant as the dataset only includes flowers of the Iris genus  
Updating the values to species name, and renaming the column to species to reflect this change

The column names also include some whitespace which is taken care of by renaming all of them

```python
# renaming columns
column_names = ["sepal_length", "sepal_width", "petal_length", "petal_width", "species"]
iris.columns = column_names

# renaming genus-species to species
species_names = {"Iris-setosa": "setosa", "Iris-versicolor": "versicolor", "Iris-virginica": "virginica"}
iris["species"] = iris["species"].map(species_names) # mapping dict to species values

iris.head()
```

Saving dataframe to csv to use in upcoming data analysis

```python
 # index = False to not save the index numbers, note that they will be re-added by default when reading csv
iris.to_csv("assets/iris_clean.csv", index = False)
```
